```python
import matplotlib.pyplot as plt
ms = 500
n_tasks = 512
subtitle = f"n_tasks = {n_tasks}, each task is {ms}ms"
target = [[1, 2, 4, 6, 8, 10, 12, 14, 16, 18, 26, 30, 36, 42, 48, 54, 60, 64], []]
target[1] = [(ms * n_tasks)/i for i in target[0]]
print(target)
```

# Standard

```python
standard = [[1, 4, 8, 16, 32, 64], [256057.643, 256087.678, 256091.556, 256095.551, 256107.402, 256127.834]]
```

```python
fig, axs = plt.subplots(figsize=(18, 5), dpi=300)
axs.grid()
# Setting the number of ticks
plt.locator_params(axis='x', nbins=len(target[0])) 
plt.locator_params(axis='y', nbins=12)
plt.plot(standard[0], standard[1], label="Standard")
plt.plot(target[0], target[1], ls="dotted", label="Target", color="grey")

axs.set_xlabel("number of workers")
axs.set_ylabel("time (msec)")
axs.set_ylim(ymin=0)

plt.suptitle('[Standard] Time over number of workers', y=0.97, fontweight="bold")
plt.title(subtitle)
plt.legend()
#savefig("stream_service_time")
plt.show()
```
