```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
import seaborn as sns
```

```python
results = [[1, 2, 4, 8, 10, 12, 16, 20, 26, 32], [0.405, 0.815, 1.632, 3.262, 4.092, 4.860, 6.552, 8.123, 10.545, 12.991]]
```

```python
fig, axs = plt.subplots(figsize=(18, 5), dpi=300)
axs.grid()
# Setting the number of ticks
plt.locator_params(axis='x', nbins=len(results[0])*2) 
plt.locator_params(axis='y', nbins=12)
plt.plot(results[0], results[1], label="Actual")
plt.plot(results[0], [0.404 for _ in results[0]], label="Expected", color="black")

axs.set_xlabel("Number of tasks")
axs.set_ylabel("Time (seconds)")

plt.title('Time over number of tasks', fontweight="bold")
plt.legend()
plt.savefig("gil_bench", bbox_inches="tight")
plt.show()
```
