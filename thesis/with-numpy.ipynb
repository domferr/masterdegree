```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams.update({'font.size': 14})
import seaborn as sns
sns.set_style("darkgrid")
#https://realpython.com/python312-subinterpreters/
```

```python
x_axis = [1, 2, 4, 8, 12, 26, 36, 48, 64]

n_tasks = 128
xticks = [1]
for i in range(5, x_axis[-1], 5):
    xticks.append(i)
if xticks[-1] != x_axis[-1]:
    xticks.append(x_axis[-1])

sequential = [63.081890448 for _ in x_axis]
fastflow = [65.505184173, 49.163296415, 41.839747256, 39.905686472, 36.907457859, 37.65657676, 37.681742395, 34.862904403, 33.432950183]
processpool = [65.512467489, 48.676692326, 44.38897567, 46.47914951, 47.983761306, 47.560968163, 41.369384828, 37.56165655, 36.697945774]
```

```python
fig, axs = plt.subplots(figsize=(18, 5), dpi=300)
colors = ["#A0D2DB", "#188FA7", "#0F5C6B", "#F25A02"]
linewidth = 2
style = '--'
plt.plot(x_axis, processpool, style, label="ProcessPoolExecutor", linewidth=linewidth, color=colors[2])[0].set_dash_capstyle('round')
plt.plot(x_axis, fastflow, '-', label="FastFlow", linewidth=linewidth, color=colors[3])[0].set_dash_capstyle('round')
plt.plot(x_axis, sequential, ls="dotted", label="Sequential", linewidth=linewidth, color="black")[0].set_dash_capstyle('round')

axs.set_xlabel("Number of workers")
axs.set_ylabel("Time (seconds)")
# Setting the number of ticks
#axs.xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(xticks, labels=xticks)
#axs.yaxis.set_major_locator(ticker.MultipleLocator(2))
axs.yaxis.set_major_formatter(ticker.ScalarFormatter())
axs.yaxis.set_minor_formatter(ticker.ScalarFormatter())
axs.tick_params(which='minor', length=0)

plt.title('Time over number of workers', fontweight="bold", y=1.05)
plt.legend()
plt.savefig("with-numpy", bbox_inches="tight")
plt.show()
```

```python
if xticks[-1] != x_axis[-1]:
    xticks.append(x_axis[-1])
fig, axs = plt.subplots(figsize=(18, 5), dpi=300)
plt.plot(x_axis, [fastflow[0]/val for val in fastflow], '-', label="FastFlow", linewidth=linewidth, color=colors[3])[0].set_dash_capstyle('round')
plt.plot(x_axis, [processpool[0]/val for val in processpool], '-', label="ProcessPoolExecutor", linewidth=linewidth, color=colors[2])[0].set_dash_capstyle('round')
plt.plot(x_axis, [nw for nw in x_axis], ls="dotted", label="Ideal", linewidth=linewidth, color="black")[0].set_dash_capstyle('round')

axs.set_xlabel("Number of workers")
axs.set_ylabel("Scalability")
# Setting the number of ticks
#axs.xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(xticks, labels=xticks)
#axs.yaxis.set_major_locator(ticker.MultipleLocator(2))
axs.yaxis.set_major_formatter(ticker.ScalarFormatter())
axs.yaxis.set_minor_formatter(ticker.ScalarFormatter())
axs.tick_params(which='minor', length=0)

plt.title('Scalability', fontweight="bold", y=1.05)
plt.legend()
#plt.savefig("with-numpy", bbox_inches="tight")
plt.show()
```
